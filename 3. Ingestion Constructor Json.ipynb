```python
constructors_schema = "constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING"


constructor_df = (spark.read \
    .schema(constructors_schema) \
    .json("abfss://bronze@mjuanworkshopetl.dfs.core.windows.net/constructors.json"))
```

```python
constructor_df.display()
```

```python
constructor_drop_df = constructor_df.drop('url')
constructor_drop_df.display()
```

```python
from pyspark.sql.functions import col, current_timestamp

constructor_renamed_df = constructor_drop_df.withColumnRenamed("constructorId", "constructor_id") \
    .withColumnRenamed("constructorRef", "constructor_ref") \
    .withColumn("ingestion_date", current_timestamp())

constructor_renamed_df.display()
```

```python
constructor_renamed_df.write.mode('overwrite').option("overwriteSchema", "true").saveAsTable('formula1_project.silver.constructors_transformed')
```
