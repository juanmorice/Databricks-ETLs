Reading tables already created in Unity Catalog using Spark

```python
races_df = spark.read.table('formula1_project.silver.races_transformed')
races_df.display()
```

```python
#using SQL syntaxis

races_filtered_df = races_df.filter("race_year = 2019 and round <= 5")
```

```python
#using Python syntaxis for &&

#races_filtered_df = races_df.filter(races_df.race_year == 2019).filter(races_df.round <= 5)

#It would be the same

races_filtered_df = races_df.filter((races_df.race_year == 2019) & (races_df.round <= 5))
```

```python
races_filtered_df = races_df.filter((races_df.race_year == 2019) | (races_df.round <= 5))
```
