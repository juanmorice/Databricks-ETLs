```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType


pitstops_schema = StructType(fields=[ 
                                     
                                     StructField("raceId", IntegerType(), True),
                                     StructField("driverId", IntegerType(), True),
                                     StructField("stop", IntegerType(), True),
                                     StructField("lap", IntegerType(), True),
                                     StructField("time", StringType(), True),
                                     StructField("duration", StringType(), True),
                                     StructField("milliseconds", IntegerType(), True)
    
    ])

pitstops_df = spark.read \
    .option("multiline", "true") \
    .schema(pitstops_schema) \
    .json("abfss://bronze@mjuanworkshopetl.dfs.core.windows.net/pit_stops.json")

pitstops_df.display()
```

```python
from pyspark.sql.functions import current_timestamp

pitstops_transf_df = pitstops_df.withColumnRenamed("raceId", "race_id") \
                                .withColumnRenamed("driverId", "driver_id") \
                                .withColumn("ingestion_date", current_timestamp())

pitstops_transf_df.display()
```

```python
pitstops_transf_df.write.mode('overwrite').option("overwriteSchema", "true").saveAsTable('formula1_project.silver.pitstops_transformed')
```
